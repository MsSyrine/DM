```python
#Import packages
import pandas as pd
import numpy as np

import json
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
#Import the datasets into dataframes
credits = pd.read_csv('tmdb_5000_credits.csv')
movies = pd.read_csv('tmdb_5000_movies.csv')
```

```python
#A Quick Look at the dataframe: Credits
credits.head()
```

```python
#A Quick Look at the dataframe: movies
movies.head()
```

```python
#Filters to limit the data and remove outliers
movies = movies[movies.original_language == 'en']#keep only original english movies
movies = movies[movies.vote_count != 0] #keep movies with non null votes
movies = movies[movies.genres != '[]']  #keep movies with non null genres
```

```python
#Clean the Dataset
#Create new dataframes from Dictionary like columns
def mapper(df, col, new_col):
    temp = []
    for n, row in df.iterrows():
        [temp.append(tuple(i.values())+(row[new_col],)) for i in json.loads(row[col])]
        
    new_df = pd.DataFrame(temp)
    new_df.columns = list(json.loads(row[col])[0].keys()) + ['movie_id']

    return new_df
```

```python
#Extracting the Data from json like columns

#extract from the credits dataset json like columns
credit_cast = mapper(credits, 'cast', 'movie_id')#cast credits
credit_crew = mapper(credits, 'crew', 'movie_id')#crew credits

#extract from the movies dataset json like columns
movie_genres = mapper(movies, 'genres', 'id') #genre
movie_keywords = mapper(movies, 'keywords', 'id') #keywords describing the movie
movie_production_companies = mapper(movies, 'production_companies', 'id') 
movie_spoken_languages = mapper(movies, 'spoken_languages', 'id')
movie_production_countries = mapper(movies, 'production_countries', 'id')
```

```python
movie_genres.columns # visualizing the movies genre columns
```

```python
#Renaming the columns for better understanding of the data fields
#renaming cast columns
credit_cast.columns = ['cast_id', 'character', 'credit_id', 'gender', 'actor_id', 'actor_name', 'order','movie_id']
#renaming crew columns
credit_crew.columns = ['credit_id', 'department', 'gender', 'job_id', 'job', 'crew_member', 'movie_id']
#renaming movie genres columns
movie_genres.columns = ['genre_id','genre','movie_id']
#renaming movie production's companies columns
movie_production_companies.columns = ['studio', 'studio_id', 'movie_id']

movie_spoken_languages.columns = ['iso_639_1', 'language', 'movie_id']#ISO 639-1 is the standard of language codes

movie_production_countries.columns = ['iso_3166_1', 'country', 'movie_id']#codes ISO 3166 des pays
```

# FIRST APPROACH

first ,we are considering in our metrics the movies popularity,revenue,runtime,vote count,vote average and the budget.
we proceed by defining the columns of the metrics so it adjust to the data contained
Secondly,after properly defining the metrics data,we perform a merge/inner join to combine the movie_genres with metrics based on movie_id (a commonly shared unique field)
Then we calculate the weights corresponding to each field within the metrics so that we can determine what movies can be considered as successful 


```python

#Creating a Weights dataframe to attach genre-wise weights to the movies,
#These weights are decided by various aspects
metrics = movies[['id','popularity','revenue','runtime','vote_count','vote_average','budget']]
#defining metrics by selecting the most important features wihin the movies dataset
```

```python
metrics.columns = ['movie_id','popularity','revenue','runtime','vote_count','vote_average','budget']#renaming metrics columns for better representation of the data contained
```

```python
genre_df = pd.merge(movie_genres, metrics, on='movie_id',how='inner') #inner join on movie_id


weights = genre_df.groupby('genre').agg({'movie_id':'count', 'revenue':'mean','runtime':'mean', 
                           'vote_count':'count','vote_average':'mean','budget':'mean'})

import seaborn as sns#for visualizing the weights of movies when it comes to rating the genre

weights = weights.reset_index().sort_values('vote_average',ascending=True)

plt.figure(figsize=(20,10))

sns.violinplot(x=genre_df['genre'], y=genre_df['vote_average'], linewidth=0.5,fontsize=15)

plt.ylabel('Rating',fontsize=13)
plt.xlabel('Genres',fontsize=13)
plt.title(' Average rating by genre ',fontsize=15)
plt.show()
```

```python

from sklearn.preprocessing import MinMaxScaler 
ss = MinMaxScaler()

metrics = movies[['id','popularity','revenue','runtime','vote_count','vote_average']]
metrics.columns = ['movie_id','popularity','revenue','runtime','vote_count','vote_average']

weights = pd.merge(movie_genres, metrics, on='movie_id',how='inner')
weights = weights.groupby('genre').agg({'movie_id':'count', 'revenue':'mean','runtime':'mean', 
                           'vote_count':'count','vote_average':'mean'})


weights_scores = np.sum(ss.fit_transform(weights),axis=1)
```

```python
def score_calculator(df1, df2, col):#this function is for calculating the score of two data entities according to a given column feature
    joined_df = pd.merge(df1, df2, on ='movie_id')
    
    #pivot tables and make the col as an index and fill the NaN values with 0
    temp = joined_df.pivot_table(index = col, columns = 'genre', values = 'genre_id', aggfunc = 'count').fillna(0)
    
    df_index = temp.index
    #calculate the score for each column property of the newly joined data element
    score_df = pd.DataFrame(np.matmul(temp.values, weights))
    score_df.index = df_index
    score_df.columns = ['score']
    score_df = score_df.sort_values(by='score',ascending=False).reset_index()
    #perform an inner join on movie_id and calculate the sum score
    score = pd.merge(df1, score_df, on = col, how = 'inner').groupby('movie_id')['score'].sum().reset_index()
    score.columns = ['movie_id',col+'_score']
    
    return score
```

```python
movies['release_month'] = movies.release_date.str.split('-').apply(lambda x:int(x[1]))
def season(row):
    if row >= 1 and row < 4:
        return 'winter'
    elif row >= 4 and row < 7:
        return 'spring'
    elif row >= 7 and row < 10:
        return 'summer'
    elif row >= 10 and row < 12:
        return 'autumn'
    else:
        return 'winter'

movies['season'] = movies.release_month.apply(season)
movies.columns
data = movies.drop(['genres', 'homepage', 'keywords', 'original_title', 'overview', 'production_companies', 
                    'release_date', 'production_countries', 'spoken_languages', 'status', 'tagline',
                    'release_month'],axis=1)
data.columns = ['budget', 'movie_id', 'original_language', 'popularity', 'revenue', 'runtime', 'title', 'vote_average',
                'vote_count', 'season']#add the season feature

#perform join operation based on the movie_id
data = pd.merge(data, cast_score, on='movie_id', how='inner')
data = pd.merge(data, production_score, on='movie_id', how='inner')
data = pd.merge(data, crew_score, on='movie_id', how='inner')

data.head()
```

```python
def plot_hist(df, col, title, x, y):
    temp = data[['title', col]].sort_values(by=col, ascending=True).head(10)#take only ten elements
    
    ax = temp.set_index('title').plot(kind='bar', figsize=(20,10), fontsize=13)
    ax.set_xlabel(x, fontsize=13)
    ax.set_ylabel(y, fontsize=13)
    ax.set_title(title, fontsize=15)
    
    plt.show()
#visualize the latest updated data with the actor name and actors movies scores
plot_hist(data, 'actor_name_score', 'Movies with Highest Actor Scores', 'Movies', 'Actor Score')
```

```python
plot_hist(data, 'crew_member_score', 'Movies with Highest Behind the Scenes Scores', 'Movies', 'Behind the Scenes Score')
```

```python
#Creating the Target Variables
#we define a threshold value for each of the vote_average,popularity and budget features where we consider 
#any movie having a lower value than this threshold as a failed movie
def create_labels(row):
    if (row['vote_average']>= 5.5)&(row['popularity']>50)&((row['revenue']-row['budget'])>0):
        return 1
    else:
        return 0

data['target'] = data[['vote_average','popularity','revenue', 'budget']].apply(create_labels,axis=1)
data.head()
```

```python
#Getting the feature and target vector
#perform a get dummies operation in order to transform the non numeric value of each season into binary values (almost boolean 1/0)
X = pd.get_dummies(data.drop(['target','movie_id','title','original_language', 
                              'budget', 'revenue', 'vote_average', 'vote_count', 'popularity'],axis=1)).values
y = data['target'].values
cols = list(pd.get_dummies(data.drop(['target','movie_id','title','original_language', 
                              'budget', 'revenue', 'vote_average', 'vote_count', 'popularity'],axis=1)).columns)
```

```python
#Divide the data into training and testing
from sklearn.model_selection import train_test_split #for partitioning data

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0, stratify=y)
#here we pass 0.3 as the value, the dataset will be split 30% as the test dataset.
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold

```

```python

# The list of [best_score_, best_params_, best_estimator_]
best_score_param_estimators = []
```

```python
#DecisionTreeClassifier
gs = GridSearchCV(estimator=Pipeline([('StandardScaler', StandardScaler()), ('clf', DecisionTreeClassifier(random_state=0))]),
                param_grid=[{'clf__min_samples_split': [2, 10, 30],
               'clf__min_samples_leaf': [1, 10, 30]}], scoring='accuracy',
                  n_jobs=-1,cv=StratifiedKFold(n_splits=10, shuffle=True, random_state=10))
# Fit the pipeline
gs.fit(X_train, y_train)
    
# Update best_score_param_estimators
best_score_param_estimators.append([gs.best_score_, gs.best_params_, gs.best_estimator_])

```

```python
#RandomForestClassifier
gs = GridSearchCV(estimator=Pipeline([('StandardScaler', StandardScaler()), ('clf', RandomForestClassifier(random_state=0))]), 
                param_grid=[{'clf__n_estimators': [2, 10, 30],
               'clf__min_samples_split': [2, 10, 30],
               'clf__min_samples_leaf': [1, 10, 30]}], scoring='accuracy', n_jobs=-1, 
                      cv=StratifiedKFold(n_splits=10, shuffle=True, random_state=10))
# Fit the pipeline
gs.fit(X_train, y_train)
    
# Update best_score_param_estimators
best_score_param_estimators.append([gs.best_score_, gs.best_params_, gs.best_estimator_])
```

```python
# Sort best_score_param_estimators in descending order of the best_score_
best_score_param_estimators = sorted(best_score_param_estimators, key=lambda x : x[0], reverse=True)

# For each [best_score_, best_params_, best_estimator_]
for best_score_param_estimator in best_score_param_estimators:
    # Print out [best_score_, best_params_, best_estimator_], where best_estimator_ is a pipeline
    # Since we only print out the type of classifier of the pipeline
    print([best_score_param_estimator[0], best_score_param_estimator[1], type(best_score_param_estimator[2].named_steps['clf'])], end='\n\n')
```

```python
from sklearn.metrics import confusion_matrix
```

```python
#RandomForestClassifier
y_pred = best_score_param_estimators[0][2].predict(X_test)
confusion_matrix(y_test, y_pred)
```

```python
#DecisionTreeClassifier

y_pred = best_score_param_estimators[1][2].predict(X_test)
confusion_matrix(y_test, y_pred)
```

```python
#  Recreate the Random Forest Classifier with the same parameters
rf = RandomForestClassifier(min_samples_leaf= 10, min_samples_split= 30, n_estimators= 30, random_state=0)
rf.fit(X_train, y_train)
rf.feature_importances_
```
